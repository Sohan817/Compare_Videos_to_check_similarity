```python
#Extract Frames from videos
import cv2
cam = cv2.VideoCapture('C:/Users/Sohan/Desktop/BPC/video2/test2.mp4')
fps = cam.get(cv2.CAP_PROP_FPS)
n = 0
i = 0
while True:
    ret,frame = cam.read()
    if(2*n)%fps == 0:
        cv2.imwrite("{}.png".format(i),frame)
        i+=1
    n+=1
    if ret == False:
        break
cam.release()
cv2.destroyAllWindows()
```

```python
#Extract data from audios
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
from glob import glob

import librosa
import librosa.display
import IPython.display as ipd
```

```python
audio_file_1 = glob('C:/Users/Sohan/Desktop/BPC/video1/audio1.mp3')
audio_file_2 = glob('C:/Users/Sohan/Desktop/BPC/video2/audio2.mp3')
```

```python
#Sample Rate
y1,sr1 = librosa.load(audio_file_1[0])
y2,sr2 = librosa.load(audio_file_2[0])
```

```python
pd.Series(y1).plot(figsize=(10,5))
```

```python
pd.Series(y2).plot(figsize=(10,5))
```

```python
pd.Series(y1).plot(figsize=(10,5))
pd.Series(y2).plot(figsize=(10,5))
```

```python
#frequency
D1 = librosa.stft(y1)
s1_db = librosa.amplitude_to_db(np.abs(D1), ref=np.max)
s1_db.shape
```

```python
fig , ax = plt.subplots(figsize=(10,5))
img = librosa.display.specshow(s1_db,xaxis='time',yaxis='log',ax=ax)
```

```python
D2 = librosa.stft(y2)
s2_db = librosa.amplitude_to_db(np.abs(D2), ref=np.max)
s2_db.shape
```
