```python
# define the function to compute MSE between two images
def mse(img_first, img_second):
   h, w = img_first.shape
   diff = cv2.subtract(img_first, img_second)
   err = np.sum(diff**2)
   mse = err/(float(h*w))
   return mse, diff
```
